## II. correlation matrices

```python
# exclude non-numeric data in correlation matrix
correlation_matrix = df.select_dtypes(include=['number']).corr()
print(correlation_matrix)
```

```python
# create a correlation heatmap using matplotlib
plt.matshow(correlation_matrix.corr())
plt.show()
```

```python
# add labels, a legend, and change the size of the heatmap

f = plt.figure(figsize=(8, 8)) # figure size 
plt.matshow(correlation_matrix.corr(), fignum=f.number) # type of plot
plt.xticks(range(correlation_matrix.shape[1]), correlation_matrix.columns, fontsize=10, rotation=90) # x axis labels
plt.yticks(range(correlation_matrix.shape[1]), correlation_matrix.columns, fontsize=10) # y axis labels
cb = plt.colorbar() # add a colour legend (called colorbar)
cb.ax.tick_params(labelsize=14) # add font size
plt.title('Correlation Matrix', fontsize=14) # add title
plt.show() # explicity display the plot
```

```python
# create a subplot with matplotlib
f,ax = plt.subplots(figsize=(10,10))

# create the correlation heatmap in seaborn 
corr = sns.heatmap(correlation_matrix.corr(), annot = True, ax = ax) # `annot` argument to place the correlation coefficients onto the heatmap
plt.show()
```

```python
# import data set free of outliers
df_no_out = pd.read_csv(os.path.join(path, 'USA Real Estate Cleaned No Outliers.csv'), index_col = False)
```

```python
# drop 'Unnamed: 0' column
df_no_out = df_no_out.drop(columns = ['Unnamed: 0'])
```

```python
# change 'brokered_by' to object/string data type
df_no_out['brokered_by'] = df_no_out['brokered_by'].astype('str')
```

```python
# change 'zip_code' to object/string data type
df_no_out['zip_code'] = df_no_out['zip_code'].astype('str')
```

```python
# exclude non-numeric data in correlation matrix 2
correlation_matrix_2 = df_no_out.select_dtypes(include=['number']).corr()
print(correlation_matrix_2)
```

```python
# create a correlation heatmap using matplotlib
plt.matshow(correlation_matrix_2.corr())
plt.show()
```

```python
# add labels, a legend, and change the size of the heatmap

f = plt.figure(figsize=(8, 8)) # figure size 
plt.matshow(correlation_matrix_2.corr(), fignum=f.number) # type of plot
plt.xticks(range(correlation_matrix_2.shape[1]), correlation_matrix_2.columns, fontsize=10, rotation=90) # x axis labels
plt.yticks(range(correlation_matrix_2.shape[1]), correlation_matrix_2.columns, fontsize=10) # y axis labels
cb = plt.colorbar() # add a colour legend (called colorbar)
cb.ax.tick_params(labelsize=14) # add font size
plt.title('Correlation Matrix', fontsize=14) # add title
plt.show() # explicity display the plot
```

```python
# create a subplot with matplotlib
f,ax = plt.subplots(figsize=(10,10))

# create the correlation heatmap in seaborn 
corr = sns.heatmap(correlation_matrix_2.corr(), annot = True, ax = ax) # `annot` argument to place the correlation coefficients onto the heatmap
plt.show()
```

These heatmaps reveal that patterns in the data made be beyond the scope of simple linear relationships. A correlation between number of beds and baths in a home is the strongest correlation and that makes sense. Additionally, it appears there is some indication that size of house, price, and broker might all be loosely correlated and may be worth further exploration. The weak correlations also point to another indicator as a large contributor to these variables (my hypothesis would be location)

## V. categorical plots

```python
# ensure status column has a categorical data type with a defined order
df_no_out['status'] = pd.Categorical(
    df_no_out['status'], 
    categories=['sold', 'ready_to_build', 'for_sale'], 
    ordered=True)

# create histogram
sns.histplot(data=df, x='status', bins=3, kde=True)

# remove scientific notation
plt.ticklabel_format(style='plain', axis='y')

# add labels and title
plt.xlabel('House Status', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.title('House Status Distribution', fontsize=14)

# show the plot
plt.show()
```

```python
df_no_out.nunique()
```

```python
# create histogram for states
sns.histplot(data=df_no_out, x='state', bins=54, kde=True)

# remove scientific notation
plt.ticklabel_format(style='plain', axis='y')
plt.xticks(fontsize=8, rotation=90)

# add labels and title
plt.xlabel('State Name', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.title('Number of Houses by State', fontsize=14)

# show the plot
plt.show()
```

this indicates number of houses likely coincides with the population of the state, perhaps merging census data might be useful
